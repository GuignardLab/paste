```python
import math
import time
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib import style
style.use('seaborn-dark')
import paste
```

## Read data and create STLayer object

```python
from paste import STLayer

def getCoordinates(df):
    """
    Extracts spatial coordinates from ST data with index in 'AxB' type format.
    
    Return: pandas dataframe of coordinates
    """
    coor = []
    for spot in df.index:
        coordinates = spot.split('x')
        coordinates = [float(i) for i in coordinates]
        coor.append(coordinates)
    return coor
```

```python
layer1_path = "sample_data/layer1.csv"
layer2_path = "sample_data/layer2.csv"
layer3_path = "sample_data/layer3.csv"
layer4_path = "sample_data/layer4.csv"
layer1_raw_df = pd.read_csv(layer1_path, index_col=0)
layer2_raw_df = pd.read_csv(layer2_path, index_col=0)
layer3_raw_df = pd.read_csv(layer3_path, index_col=0)
layer4_raw_df = pd.read_csv(layer4_path, index_col=0)


layer1 = STLayer(layer1_raw_df, getCoordinates(layer1_raw_df))
layer2 = STLayer(layer2_raw_df, getCoordinates(layer2_raw_df))
layer3 = STLayer(layer3_raw_df, getCoordinates(layer3_raw_df))
layer4 = STLayer(layer4_raw_df, getCoordinates(layer4_raw_df))
```

Preprocess gene expression matrices

```python
layer1.filter_genes(15)
layer2.filter_genes(15)
layer3.filter_genes(15)
layer4.filter_genes(15)
```

Each STLayer consists of a gene expression matrix (pandas dataframe) and spatial coordinate matrix (numpy array).

Note, you can choose to label the spots however you want. In this case, we use the default coordinates.

# Pairwise Alignment

### Sequential pairwise layer alignment plots

```python
from paste import stack_layers_pairwise

pis = [pi12, pi23, pi34]
layers = [layer1, layer2, layer3, layer4]

new_layers = stack_layers_pairwise(layers, pis)
```

Now that we've aligned the spatial coordinates, we can plot them all on the same coordinate system.

```python
layer_colors = ['#e41a1c','#377eb8','#4daf4a','#984ea3']

plt.figure(figsize=(7,7))
for i in range(len(new_layers)):
    plot_layer(new_layers[i],layer_colors[i],s=400)
plt.legend(handles=[mpatches.Patch(color=layer_colors[0], label='1'),mpatches.Patch(color=layer_colors[1], label='2'),mpatches.Patch(color=layer_colors[2], label='3'),mpatches.Patch(color=layer_colors[3], label='4')])
plt.gca().invert_yaxis()
plt.axis('off')
plt.show()
```

We can also plot pairwise layers together.

```python
layer_colors = ['#e41a1c','#377eb8','#4daf4a','#984ea3']

fig, axs = plt.subplots(2, 2,figsize=(7,7))
plot_layer(new_layers[0],layer_colors[0],ax=axs[0,0])
plot_layer(new_layers[1],layer_colors[1],ax=axs[0,0])
plot_layer(new_layers[1],layer_colors[1],ax=axs[0,1])
plot_layer(new_layers[2],layer_colors[2],ax=axs[0,1])
plot_layer(new_layers[2],layer_colors[2],ax=axs[1,0])
plot_layer(new_layers[3],layer_colors[3],ax=axs[1,0])
fig.delaxes(axs[1,1])
plt.show()
```

We can also plot the layers in 3-D.

```python
import plotly.express as px
import plotly.io as pio
pio.renderers.default='notebook'

layer_colors = ['#e41a1c','#377eb8','#4daf4a','#984ea3']

# scale the distance between layers
z_scale = 2

df = pd.DataFrame(columns=['x','y','z','layer'])
for i,L in enumerate(new_layers):
    for x,y in L.coordinates:
        df = df.append({'x': x,'y':y,'z':i*z_scale,'layer':str(i)}, ignore_index=True)
fig = px.scatter_3d(df, x='x', y='y', z='z',
              color='layer',color_discrete_sequence=layer_colors)
fig.update_layout(scene_aspectmode='data')
fig.show()
```

# Center Alignment

First, we will read in and preprocess the data (if you ran `pairwise_align` above, it will be altered).

```python
layer1_path = "sample_data/layer1.csv"
layer2_path = "sample_data/layer2.csv"
layer3_path = "sample_data/layer3.csv"
layer4_path = "sample_data/layer4.csv"
layer1_raw_df = pd.read_csv(layer1_path, index_col=0)
layer2_raw_df = pd.read_csv(layer2_path, index_col=0)
layer3_raw_df = pd.read_csv(layer3_path, index_col=0)
layer4_raw_df = pd.read_csv(layer4_path, index_col=0)


layer1 = STLayer(layer1_raw_df, getCoordinates(layer1_raw_df))
layer2 = STLayer(layer2_raw_df, getCoordinates(layer2_raw_df))
layer3 = STLayer(layer3_raw_df, getCoordinates(layer3_raw_df))
layer4 = STLayer(layer4_raw_df, getCoordinates(layer4_raw_df))

layer1.filter_genes(15)
layer2.filter_genes(15)
layer3.filter_genes(15)
layer4.filter_genes(15)
```

### Run PASTE `center_align`.

```python
layers = [layer1, layer2, layer3, layer4]
initial_layer = layer1
lmbda = len(layers)*[1/len(layers)]
```

```python
start = time.time()

W, H, pis = center_align(initial_layer, layers, lmbda)

print('Runtime: ' + str(time.time() - start))
```

`center_align` returns the low dimensional returns the low dimensional representation of our inferred center layer.

```python
pd.DataFrame(W)
```

```python
pd.DataFrame(H)
```

However, we can easily convert the low dimensional representation back into an STLayer object if we choose.

```python
center_layer = STLayer(pd.DataFrame(np.dot(W, H), columns = initial_layer.gene_exp.columns), initial_layer.coordinates)
```

### Center layer alignment plots

Next, we can use the outputs of `center_align` to align the layers.

```python
center, new_layers = stack_layers_center(center_layer, layers, pis)
```

Now that we've aligned the spatial coordinates, we can plot them all on the same coordinate system. Note the center layer is not plotted.

```python
center_color = 'orange'
layer_colors = ['#e41a1c','#377eb8','#4daf4a','#984ea3']

plt.figure(figsize=(7,7))
plot_layer(center,center_color,s=400)
for i in range(len(new_layers)):
    plot_layer(new_layers[i],layer_colors[i],s=400)

plt.legend(handles=[mpatches.Patch(color=layer_colors[0], label='1'),mpatches.Patch(color=layer_colors[1], label='2'),mpatches.Patch(color=layer_colors[2], label='3'),mpatches.Patch(color=layer_colors[3], label='4')])
plt.gca().invert_yaxis()
plt.axis('off')
plt.show()
```
